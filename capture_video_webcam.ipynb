```python
import cv2,time
video=cv2.VideoCapture(0)

a=1
while True:
    a=a+1
    check,frame=video.read()
    print(frame)

#     frame=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)   #     print gray frames, add below line
    frame = cv2.flip(frame,1)     # flipping(mirror) frames
    cv2.imshow("Capture",frame)
    key=cv2.waitKey(1)
    if key== ord('q'):
        break

print(a)
video.release()
cv2.destroyAllWindows()


```
